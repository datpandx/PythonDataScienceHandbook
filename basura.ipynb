```python
# counting rainy days
rainfall = pd.read_csv('Seattle2014.csv')['PRCP'].values
inches= rainfall / 254 #1/10 mm -> inches
inches.shape
```

```python
plt.hist(inches,25)
```

```python
np.sum((inches>0.5)&(inches<1))
```

```python
print("# days w/o rain:               ", np.sum(inches==0))
print("# days w/ rain:                ", np.sum(inches!=0))
print("Days with more than 0.5 inches:", np.sum(inches>0.5))
print("rainy days w/ <0.1 inches:     ", np.sum((inches>0.0)&(inches<0.2)))
```

```python
# boolean operation as an index x[x>n]

# mask of all rainy days
rainy = (inches>0)

#summer days (june 21st is the 172nd of the year)
summer = (np.arange(365)-172 < 90)& (np.arange(365)-172 >0)
```

```python
print("Median precip on rainy days in 2014 (inches):  ",np.median(inches[rainy]))
print("Median precip on summer days in 2014 (inches): ",np.median(inches[summer]))
print("Maximum precip on summer days in 2014 (inches): ",np.max(inches[summer]))
print("Maximum precip on summer days in 2014 (inches): ",np.max(inches[rainy & summer]))
```

```python
def f(x,y):
    return np.sin(x)**10+np.cos(10+y*x)*np.cos(x)
x = np.linspace(0,5,50)
y = np.linspace(0,5,40)
```

```python
X, Y = np.meshgrid(x,y)
Z = f(X,Y)
```

```python
plt.contour(X, Y, Z, cmap='RdGy')
plt.colorbar()

```

```python
plt.contourf(X, Y, Z, cmap='RdGy')
plt.colorbar()
```
